```python
import pandas as pd  #read csv file (stock market data is in csv file )
import numpy as np 
import matplotlib.pyplot as plt #visualize the data
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Activation
import datetime
```

```python
''' read stock market data'''
vs_data = pd.read_csv('GOOG.csv') 

high = vs_data['High'].values #bring the data of high prices 
low = vs_data['Low'].values #bring the data of low prices
mid  = (high+low)/2 #mid prices -> we are just predicting the low prices
```

```python
'''create windows for the data of the 51 days
this store the data of 50 days and predict the data of the next date'''

seq_len = 30 #data set of 30 days
sequence_length = seq_len + 1 #data set of 30 + 1 days (stored + predicted)

result = [] 
for index in range (len(mid) - sequence_length):
    result.append(mid[index:index + sequence_length]) #append every windows, which includes the data of 30days and the data of the next day
```

```python
result
```

```python
for window in result:
    print (window)
```

```python
'''nomalize the data
let the first window be 0 by dividing by itself and -1
set the relative value for the others
by dividing its data by the first window and - 1'''

normalized_data = []
for window in result:
    normalized_window = [(float(p)/float(window[0])-1)for p in window]
    normalized_data.append (normalized_window)

result = np.array(normalized_data)
```

```python
result
```

```python
size = int(round(result.shape[0]*0.9)) #90% of data = 1084
```

```python
'''training data'''
train = result [:size,:] #90% of data = training data
np.random.shuffle (train) #shuffle the training data

train_50days = train[:,:-1]  #data of 50 days
train_50days = np.reshape (train_50days,(train_50days.shape[0],train_50days.shape[1],1))
train_1day = train[:,-1] #data of next 1 day
```

```python
'''testing data'''
test_50days = result[size:,:-1]  #10% of data = testing data
test_50days = np.reshape (test_50days,(test_50days.shape[0],test_50days.shape[1],1))
test_1day = result[size:,-1] #data of next 1 day
```

```python
train_50days.shape, test_50days.shape #1084일의 데이터를 학습해서 120일의 주식 가격을 예측
```

```python
'''builind a mondel'''
model = Sequential() #모델을 순차적으로 정의 하는 클래스
model.add(LSTM(30, return_sequences = True, input_shape = (30,1))) #input = 50 개
model.add(LSTM(50, return_sequences = False)) #조정가능
model.add(Dense(1,activation = "linear")) #ouput = 1개, 다음날 하루의 데이터를 예측한다
model.compile (loss = "mse",optimizer = "rmsprop") #손실함수 = mean square error
model.summary() #50개가 들어가서 한개가 나온다 
```

```python
'''training'''
model.fit (train_50days, train_1day, 
          validation_data = (test_50days,test_1day),
           batch_size = 10, #10개씩 묶어서 학습
           epochs = 20) #number of repetition
```

```python
pred = model.predict(test_50days)
fig = plt.figure (facecolor = "white")
ax = fig.add_subplot(111)
ax.plot (test_1day, label = "original")
ax.plot (pred,label = "prediction")
ax.legend()
plt.show()
```
