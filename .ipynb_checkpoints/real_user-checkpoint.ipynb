```python
test_1_day, test_x_days, predict, test_date_start, test_date_end, test_date = LSTM.stock_analyzer(user.company, user.CO, user.date, user.LSTM_date)
```

```python
sentiment = Senti.sentiment_analyzer(user.company, test_date_start, test_date_end, test_date)
```

```python
'''user can just run the file to start data analysis and visualisation'''

'''Data Visulization'''
fit = ((max(test_1_day)+max(predict)[0])/2)-((min(test_1_day)+min(predict)[0])/2)
fit_sentiment = []
for i in sentiment: # compressed the sentiment data to fit nicely with stock value
    fit_sentiment.append(i*fit)
    
figure = plt.figure (facecolor = "white")
ax = figure.add_subplot(1,1,1)
ax.plot (test_1_day, label = "original stock data")
ax.plot (predict,label = " predicted stock data")
ax.plot (fit_sentiment,label = "sentiment score") # sentiment data compressed to nicely fit with stock data
ax.set_xlabel('total ' + str(user.date) + ' days (non-consecutive)')
ax.set_ylabel('nomralized stock & sentiment value')
ax.set_title(user.company + 
             ' ('+ str(test_date_start) + ' to ' + str(test_date_end) +')' + 
             '\n' + LSTM.title + ' Stock Values' + 
             '\n' + '& Sentiment Analysis' )
ax.legend()
plt.show()
```
